# Groq API Key Test Notebook

This notebook helps you test your **Groq API key** using the official `groq` Python client. It will verify that your API key is set up correctly and perform a quick inference test using the **llama-3.1-8b-instant** model.

```python
# Install the necessary libraries
!pip install groq python-dotenv
```

```python
import os
from dotenv import load_dotenv

# Load environment variables from the .env file
load_dotenv()

groq_key = os.getenv("GROQ_API_KEY")

if not groq_key:
    print("🔑 GROQ_API_KEY not found in environment or .env file.")
    # Prompt the user to enter the key
    groq_key = input("Please enter your Groq API Key (starts with gsk_): ").strip()
    if groq_key:
        # Write to .env file
        with open(".env", "a") as f:
            f.write(f"\nGROQ_API_KEY={groq_key}\n")
        print("✅ GROQ_API_KEY saved to .env file!")
        # Set in environment variables for this session
        os.environ["GROQ_API_KEY"] = groq_key
    else:
        print("❌ No API key entered. Cannot proceed with the API test.")
else:
    print("✅ GROQ_API_KEY detected in environment or .env file.")
```

```python
# Run the API test using the Groq Python Client
if os.environ.get("GROQ_API_KEY"):
    from groq import Groq
    
    # Initialize the Groq client (loads API key from os.environ by default)
    client = Groq()
    
    print("🚀 Sending request to Groq API (model: llama-3.1-8b-instant)...\n")
    try:
        completion = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {
                    "role": "system",
                    "content": "You are a helpful and concise assistant."
                },
                {
                    "role": "user",
                     "content": "Explain Groq and its LPU technology in one sentence."
                }
            ],
            temperature=0.7,
            max_tokens=100
        )
        print("🎉 Success! Response from Llama-3.1 via Groq:")
        print("-" * 60)
        print(completion.choices[0].message.content)
        print("-" * 60)
    except Exception as e:
        print(f"❌ Error calling Groq API: {e}")
        print("Please check if your API key is valid or if you are being rate-limited.")
else:
    print("❌ GROQ_API_KEY is not set. Please set it before running this cell.")
```
