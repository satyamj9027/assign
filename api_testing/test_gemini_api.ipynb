# Google GenAI SDK Test Script

This notebook demonstrates how to load environment variables from a `.env` file and use the new `google-genai` SDK to make a simple request to Gemini 2.5 Flash.

```python
# Install the necessary libraries (uses --no-warn-script-location to suppress PATH warnings)
!pip install google-genai python-dotenv -q
```

```python
import os
from dotenv import load_dotenv
from google import genai

# Load environmental variables from the .env file
load_dotenv()

# Verify API Key is present
api_key = os.getenv("GOOGLE_API_KEY")
if not api_key:
    print("❌ ERROR: GOOGLE_API_KEY is not set. Please update the .env file.")
else:
    print("✅ GOOGLE_API_KEY detected successfully.")
```

```python
# Initialize client and generate content
if api_key:
    # Initialize Google GenAI client with the specified GOOGLE_API_KEY
    client = genai.Client(api_key=api_key)
    
    print("Sending request to Gemini 2.5 Flash...")
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents='Explain quantum computing in one sentence.'
    )
    print("\nResponse from Gemini:")
    print(response.text)
else:
    print("Please set the GOOGLE_API_KEY before running this cell.")
```
